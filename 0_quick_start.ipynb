# Example 1. Solving a SAT instance (`.cnf`) with CNLM-Langevin solver

```python
import pycnlm

result = pycnlm.solve_sat_file("examples/sample.cnf")
print(result)
```

```python
from pycnlm import HOBO
from pycnlm.core.HOBOReducers import PTR_Ishikawa, FERQ

# E(x) = x0·x1·x2·x3
poly = {frozenset({0, 1, 2, 3}): 1.0}
hobo = HOBO(poly)

reducer_1 = PTR_Ishikawa()
result_1  = reducer_1.apply(hobo)
reducer_2 = FERQ()
result_2 = reducer_2.apply(hobo)

print(result_1)
print(result_2)
```
